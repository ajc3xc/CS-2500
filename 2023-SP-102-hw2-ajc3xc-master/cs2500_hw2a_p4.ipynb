```python
def sortByFrequency(arr):
    #initialize variables
    values_array=[]
    frequency_array = []
    array_length = len(arr)
    max_element=0
    
    #find k (the max element in the array)
    for i in range(array_length):
        if arr[i] > max_element:
            max_element = i
    
    #instantiate variables (set frequency to 0, values from 0-max_element+1)
    values_array = [i for i in range(max_element-1)]
    frequency_array = [0 for i in range(max_element-1)]
    
    #get the frequencies for the values
    for i in range(array_length):
        frequency_array[arr[i]] += 1
    #reset array
    arr.clear()
    
    #sort the frequencies using insertion sort
    #this obviously could be optimized, but it's convenient
    for i in range (0, max_element-1):
        for j in range(i+1, max_element-1):
            if frequency_array[j] > frequency_array[i]:
                frequency_array[i], frequency_array[j] = frequency_array[j], frequency_array[i]
                values_array[i], values_array[j] = values_array[j], values_array[i]
        for k in range(frequency_array[i]):
            arr.append(values_array[i])
                
    
 
 
# Driver code
arr = [3, 5, 2, 1, 0, 1, 2, 3, 4, 2, 0, 3, 4, 2, 1]


print("Unsorted array:")
print(arr)
print("")

# Function call
sortByFrequency(arr)

print("Sorted array:")
print(arr)
```
