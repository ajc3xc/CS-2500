### Imports

```python
from random import randint
from copy import deepcopy
```

### Job Class

```python
class Job:
    def __init__(self, start, finish, value):
        self.start = start
        self.finish = finish
        self.value = value
```

### Job List Sorter

```python
def Sort_Jobs(arr):
    arr.sort(key=lambda x: x.finish)
```

```python
def Greedy_Sort_Jobs(arr):
    sorted(arr, key=lambda x: (x.value)/(x.finish-x.start), reverse=True)
```

### Dynamic Programming Function

```python
#Job list is assumed to be sorted by finish time (ascending)
def Dynamic_Task_Selection(job_list):
    n = len(job_list)
    total_profit = [0]*10
    total_profit[n-1]=job_list[n-1].value
    
    for i in reversed(range(n-1)):
        compatible_table=[0]
        for j in range(i+1, n):
            if job_list[i].finish <= job_list[j].start:
                compatible_table.append(job_list[j].value)
        job_list[i].value += max(compatible_table)
        total_profit[i] = job_list[i].value
    return max(total_profit)
        
        
```

### Greedy Function

```python
def Greedy_Task_Selection(job_list):               
    total_profit = 0
    memoize = [0]*(2*n + 1)
    for i in range(0,n):
        count = 0
        for j in range(job_list[i].start, job_list[i].finish):
            count += memoize[j]
        if count == 0:
            for k in range(job_list[i].start, job_list[i].finish):
                memoize[k] = 1
            total_profit += job_list[i].value
        #print(m)  //for testing
    return(total_profit)   
```

### Main Function

```python
#set size of tasks, and amount of loops to run
n, loops = 10, 100
#to help create a limit on the values
max_finish, max_value = 15, 200

#get average run time
greedy_value_list=[]
dynamic_value_list=[]
for i in range(loops):
    
    #create jobs
    job_list=[]
    for i in range (n):
        finish=randint(1,max_finish)
        start = randint(0, finish-1)
        value = randint(1, max_value)
        #print(str(start)+" "+str(finish)+" "+str(value))
        job_list.append(Job(start, finish, value))
        #create copy of job list
        greedy_job_list=deepcopy(job_list)
    
    #sort the jobs
    Sort_Jobs(job_list)
    Greedy_Sort_Jobs(greedy_job_list)
    
    greedy_value_list.append(Greedy_Task_Selection(greedy_job_list))
    dynamic_value_list.append(Dynamic_Task_Selection(job_list))

print("The average greedy value was: " + str(sum(greedy_value_list)/loops))
print("The average dynamic value was: "+ str(sum(dynamic_value_list)/loops))

```
