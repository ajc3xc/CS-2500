### Imports

```python
#allows easier way to creating dictionary of lists
from collections import defaultdict
```

### Graph Class

```python
class Graph:
    #create dictionary of lists
    def __init__(self):
        self.graph = defaultdict(list)
    
    #create single edge
    def add_edge(self, u, v):
        self.graph[u].append(v)
    
    #add multiple edges at once (For Problem 1 only)
    #since graph isn't directed, both are added to each other's list
    #input: tuple/list of lists/tuples
    def add_edges(self, list_lists):
        for i in range (len(list_lists)-1):
            for i1 in list_lists[i]:
                for i2 in list_lists[i+1]:
                    self.add_edge(i1, i2)
                    #print(str(i1)+" "+str(i2))
                    self.add_edge(i2, i1)
                    #print(str(i2)+" "+str(i1))
```

### Breadth First Search

```python
def Breadth_First_Search(Graph, start):
        # Make list of vertices explored
        visited = [False] * (max(Graph.graph) + 1)
 
        # Create a queue for BFS
        queue = []
 
        # Mark the source node as
        # visited and enqueue it
        queue.append(start)
        visited[start] = True
 
        while queue:
 
            # Dequeue a vertex from
            # 1st index, print it
            start = queue.pop(0)
            print(start, end=" popped, ")
 
            # Get all adjacent vertices of the
            # dequeued vertex s. If a adjacent
            # has not been visited, then mark it
            # visited and enqueue it
            for i in Graph.graph[start]:
                if visited[i] == False:
                    queue.append(i)
                    visited[i] = True
                    print(str(i)+" appended, ", end=" ")
            print("")
```

### Depth First Search

```python
def Depth_First_Search(Graph, start):
    # Make list of vertices explored
    visited = [False] * (max(Graph.graph) + 1)
    
    #Create a stack for DFS
    stack = []
    
    #append start to stack
    stack.append(start)
    print(str(start)+" appended, ", end=" ")
    
    while stack:
        current = stack.pop()
        
        if not visited[current]:
            stack.append(current)
            visited[current] = True
            for i in Graph.graph[current]:
                if visited[i] == False:
                    stack.append(i)
                    print(str(i)+" appended, ", end=" ")
                    break
        else:
            print(str(current)+ " popped")            
```

### Main Function

```python
#generate graph
graph = Graph()
edge_list = ([1], [2, 3], [4, 5, 6], [7, 8], [9])
graph.add_edges(edge_list)
print("Graph generated")
print("")

print("Breadth First Search")
Breadth_First_Search(graph, 1)
print("")

print("Depth First Search")
Depth_First_Search(graph, 1)
print("")

print("Graph searched")
```
